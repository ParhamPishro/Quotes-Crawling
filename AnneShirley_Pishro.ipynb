# Web Scraping

Import Libraries

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup 
import re
import numpy as np
```

Load Data

```python
url= "https://bookroo.com/quotes/anne-of-green-gables"
page = urlopen(url)
page
```

```python
html_bytes = page.read()
html = html_bytes.decode("utf-8")
print(html)
```

Create a BeautifulSoup Object

```python
soup = BeautifulSoup(html, "html.parser")
print(soup.get_text())
```

Find Quotes

```python
div_list = soup.find_all("div")
quotes = []

for div in div_list:
  classes = div.attrs.get("class")

  if (classes == ["css-eez2cp"]):
    quote = re.sub("<.*?>", "", str(div))
    quotes.append(quote)
quotes
```

Find Quoters

```python
a_classes = []
quoters = []
characters = set()
a_list = soup.find_all("a")

for a in a_list:
  a_classes.append(a.attrs.get("class"))

for i in range(len(a_classes)):
  quoter = []
  if a_classes[i] != None:
    if (a_classes[i][1] == 'css-1lzim2z'):
      for j in range(2):
      #quoter = re.sub("<.*?>", "", str(a_list[i+1]))
        if (a_classes[i+j+1][1] == 'css-1s1o1fc'):
          quoter.append(re.sub("<.*?>", "", str(a_list[i+j+1])))
          characters.add(re.sub("<.*?>", "", str(a_list[i+j+1])))
        elif quoter == []:
          quoter.append("Nobody")
          characters.add("Nobody")
  if quoter != []:
    quoters.append(quoter)

print(quoters)
print(characters)
```

Create Dictionary

```python
MyDict = {}

for character in characters:
  for i in range(len(quotes)):
    context = quotes[i]
    if character in quoters[i]:
      if character in MyDict.keys():
        MyDict[character] = MyDict[character] + context
      else:
        MyDict[character] = context

MyDict
```
